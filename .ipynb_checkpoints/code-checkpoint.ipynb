```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm

df = pd.read_csv("data.csv")
df
```

```python
df["RC"] = df.apply(lambda row:
    row["R"] * row["C"], axis = 1)
df.plot(x='RC', y='f', kind='scatter', color='blue', marker='o', label='Data Points')

# Set plot labels and title
plt.xlabel('RC')
plt.ylabel('f')
plt.title('f en fonction de RC')
plt.show()
```

```python
df["1/f"] = df.apply(lambda row:
    row["f"]**-1, axis = 1)

df.plot(x='RC', y='1/f', kind='scatter', color='blue', marker='o')

# Set plot labels and title
plt.xlabel('$RC$ [$\Omega$]')
plt.ylabel('f [Hz]')
plt.title('$1/f$ en fonction de $RC$')
plt.show()
```

```python
df["RC_uncertainty"] = 0 # négligeable
df["f_uncertainty"] = df["f"] * 0.05


X = df['RC'].values
Y = df['1/f'].values
X_uncertainty = df['RC_uncertainty'].values
Y_uncertainty = df['f_uncertainty'].values

X_with_intercept = sm.add_constant(X)  # Add a constant term (intercept) to the X values

# Weighted least squares regression
model = sm.WLS(Y, X_with_intercept, weights=1 / Y_uncertainty**2)
results = model.fit()

# Print the regression summary
print(results.summary())

slope = results.params[1]
intercept = results.params[0]

standard_errors = results.bse

# The standard error for the slope (coefficient of X) is typically the second element in the standard_errors array
slope_standard_error = standard_errors[1]

print("Standard Error of the Slope:", slope_standard_error)

predicted_Y = results.predict(X_with_intercept)
plt.scatter(X, Y, label='1/f')
plt.plot(X, predicted_Y, color='red', label='Régression linéaire')

plt.xlabel('$RC$ [$\Omega F$]')
plt.ylabel('f [Hz]')
plt.title('$1/f$ en fonction de $RC$')

plt.legend()
plt.show()
```
